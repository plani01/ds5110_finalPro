```python
#Trial Set Up
from faker import Faker
import pandas as pd
faker = Faker(locale = 'en_US')
def generateData(num):
    movieInformation =[]
    for i in range(num):
        MovieInfo = {}
        MovieInfo['Release Year'] = faker.random_int(min = 2000,max = 2024)
        MovieInfo['Release Month'] = faker.random_element(elements=("January","Feburary","March","April","May","June","July","August","September","October","November","Decemeber"))
        MovieInfo['Title'] = faker.catch_phrase()
        MovieInfo['Rating'] = faker.random_element(elements=("G","PG","PG-13","R","M"))
        MovieInfo['Duration Time'] = faker.random_int(min = 60, max = 180)
        MovieInfo["Production Company"] = faker.random_element(elements=("Paramount Pictures","Walt Disney Company","Warner Bros","Universal Pictures","20th Century Studios","MGM Studios","Sony Pictures Entertainment","Lionsgate","DreamWorks","Columbia Pictures","Apple Studios","Gaumont","Harpo Productions","Cinemax films","Lucasfilm","Legendary Entertainment"))
        MovieInfo['Director'] = faker.name()
        MovieInfo['Genre'] = faker.random_element(elements=("Action","Comedy","Drama","Horror","Sci-Fi","Fantasy","Romance","Thriller","Mystery"))
        MovieInfo['Box Office Length (Weeks)'] = faker.random_int(min = 4, max = 12) 
        MovieInfo['Budget'] = faker.random_int(min = 10000000, max = 200000000)
        MovieInfo['Revenue'] = faker.random_int(min = 10000000, max = 500000000)
        MovieInfo['Origin of Production'] = faker.state()
        MovieInfo['Rotten Tomatoes Score'] = faker.random_int(min = 0, max = 100)
        MovieInfo['International Release'] = faker.random_element(elements=("Yes","No"))
        movieInformation.append(MovieInfo)
    return pd.DataFrame(movieInformation)
movieNames = generateData(100000)
print(movieNames)
```

```python
#Concession Table Generation
from faker import Faker
from datetime import datetime
import pandas as pd
faker = Faker(locale = 'en_US')
def generateData(num):
    numbers = [
    1243, 1115, 1152, 1973, 1679, 1592, 1934, 1125, 1208, 1943, 
    1803, 1648, 1975, 1347, 1155, 1645, 1429, 1014, 1791, 1800, 
    1419, 1546, 1610, 1698, 1810, 1632, 1211, 1093, 1079, 1082, 
    1600, 1767, 1651, 1554, 1832, 1494]
    movieInformation =[]
    for i in range(num):
        MovieInfo = {}
        MovieInfo['Concession Type'] = faker.random_element(elements = ("popcorn","soda","candy"))
        if MovieInfo['Concession Type'] == 'popcorn':
            MovieInfo['Concession ID'] = 200
            MovieInfo['Price of Item'] = 7
        if MovieInfo['Concession Type'] == 'soda':
            MovieInfo['Concession ID'] = 201
            MovieInfo['Price of Item'] = 6
        if MovieInfo['Concession Type'] == 'candy':
            MovieInfo['Concession ID'] = 202
            MovieInfo['Price of Item'] = 4
        MovieInfo["Amount Purchased"] = faker.random_int(min = 1, max = 6)
        start_date = datetime.strptime('2023-01-01', '%Y-%m-%d').date()
        end_date = datetime.strptime('2024-01-01', '%Y-%m-%d').date()   
        MovieInfo['Purchased Date'] = faker.date_between(start_date=start_date, end_date=end_date)
        MovieInfo['Employee ID'] = faker.random_element(elements = numbers)
        movieInformation.append(MovieInfo)
    return pd.DataFrame(movieInformation)
movieNames = generateData(45000)
movieNames.to_csv("ConcessionData.csv", index = False)
```

```python
#Customer Table
import numpy as np
from faker import Faker
import pandas as pd
faker = Faker(locale = 'en_US')
def generateData(num):
    movieInformation = []
    customer_id = 4000
    mean = 25
    std_dev = 10
    for i in range(num):
        MovieInfo = {}
        MovieInfo['Customer ID'] = customer_id
        MovieInfo['Customer Name'] = faker.name()
        MovieInfo['Gender'] = faker.random_element(elements = ("M","F"))
        age = int(np.random.normal(loc=mean, scale=std_dev)) 
        age = np.clip(age, 2, 90)
        MovieInfo['Age'] = age
        customer_id += 1
        
        movieInformation.append(MovieInfo)
    
    return pd.DataFrame(movieInformation)
movieNames = generateData(20000)
movieNames.to_csv("CustomerData.csv", index = False)
```

```python
#Ticket Table Generation
from faker import Faker
from datetime import datetime
import random
import pandas as pd
faker = Faker(locale = 'en_US')
types = ["standard", "3D", "VIP"]
weights = [0.6, 0.2, 0.1]
numbers = [
    1952, 1107, 1717, 1115, 1384, 1833, 1216, 1098, 1027, 1440,
    1641, 1296, 1262, 1666, 1872, 1174, 1967, 1344, 1325, 1121,
    1332, 1419, 1621, 1279, 1022, 1703, 1803, 1096, 1159, 1175,
    1482, 1169, 1246, 1239, 1587, 1267, 1838, 1546, 1797, 1087,
    1602, 1326, 1060, 1354, 1541, 1874, 1115, 1727, 1230, 1137,
    1419, 1962
]

def generateData(num):
    movieInformation = []
    reviewNum = 1
    for i in range(num):
        MovieInfo = {}
        MovieInfo["Ticket ID"] = reviewNum
        promoStart = datetime.strptime('2023-10-01', '%Y-%m-%d').date()
        promoEnd = datetime.strptime('2023-12-31', '%Y-%m-%d').date()
        start_date = datetime.strptime('2023-01-01', '%Y-%m-%d').date()
        end_date = datetime.strptime('2024-01-01', '%Y-%m-%d').date() 
        MovieInfo['Purchased Date'] =  faker.date_between(start_date=start_date, end_date=end_date)
        MovieInfo['Category'] = random.choices(types, weights=weights, k=1)[0]
        if MovieInfo['Category'] == 'standard':
            MovieInfo['Price of Ticket'] = 12
        if MovieInfo['Category'] == '3D':
            MovieInfo['Price of Ticket'] = 16
        if MovieInfo['Category'] == 'VIP':
            MovieInfo['Price of Ticket'] = 24
        MovieInfo["Tickets Sold"] = faker.random_int(min = 1, max= 6)
        if promoStart <= MovieInfo['Purchased Date'] < promoEnd:
            MovieInfo['Promotion'] = 0.85
        else:
            MovieInfo['Promotion'] = 1.00
        MovieInfo["Discount"] = faker.random_int(min = 0, max = 10)
        MovieInfo['Employee ID'] = faker.random_element(elements = numbers)
        MovieInfo['Customer ID'] = faker.random_element(elements = customerList)
        movieInformation.append(MovieInfo)
        reviewNum += 1
    return pd.DataFrame(movieInformation)
movieNames = generateData(30000)
movieNames.to_csv("Tickets.csv", index = False)

customersPurchased = movieNames['Customer ID']
print(customersPurchased)

```

```python
#ShowTime Table
from faker import Faker
import datetime
from datetime import datetime,timedelta
from random import randint
import pandas as pd
faker = Faker(locale = 'en_US')
def generateData(num):
    Status = ["On time", "Canceled", "Rescheduled"]
    weights = [0.7, 0.2, 0.1]
    movieInformation = []
    reviewNum = 1
    for i in range(num):
        randomHour = randint(9,23)
        randomMinute = randint(0,59)
        MovieInfo = {}
        MovieInfo['ShowTime ID'] = reviewNum
        MovieInfo['Screen Number'] = faker.random_int(min = 1, max = 5)
        MovieInfo['Movie Time'] = f"{randomHour:02}:{randomMinute:02}"
        MovieInfo['Showtime Status'] = random.choices(Status, weights=weights, k=1)[0]
        if MovieInfo['Showtime Status'] == "Canceled":
            MovieInfo["Reason of Status"] = faker.random_element(elements=("Screen Broken","No Tickets Purchased","Other"))
        elif MovieInfo['Showtime Status'] == "Rescheduled":
            MovieInfo["Reason of Status"] = faker.random_element(elements=("theater needs cleaning","Seats Broken","Other"))
        else:
            MovieInfo["Reason of Status"] = "On Time"
        MovieInfo["Movie ID"] = faker.random_int(min = 1, max = 30)
        MovieInfo['Customer ID'] = faker.random_element(elements = customersPurchased)
        reviewNum += 1
        movieInformation.append(MovieInfo)
    return pd.DataFrame(movieInformation)

movieNames = generateData(10000)
movieNames.to_csv("ShowTimeData.csv", index = False)
    




```

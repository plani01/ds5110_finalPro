```python
#Trial Set Up
from faker import Faker
import pandas as pd
faker = Faker(locale = 'en_US')
def generateData(num):
    movieInformation =[]
    for i in range(num):
        MovieInfo = {}
        MovieInfo['Release Year'] = faker.random_int(min = 2000,max = 2024)
        MovieInfo['Release Month'] = faker.random_element(elements=("January","Feburary","March","April","May","June","July","August","September","October","November","Decemeber"))
        MovieInfo['Title'] = faker.catch_phrase()
        MovieInfo['Rating'] = faker.random_element(elements=("G","PG","PG-13","R","M"))
        MovieInfo['Duration Time'] = faker.random_int(min = 60, max = 180)
        MovieInfo["Production Company"] = faker.random_element(elements=("Paramount Pictures","Walt Disney Company","Warner Bros","Universal Pictures","20th Century Studios","MGM Studios","Sony Pictures Entertainment","Lionsgate","DreamWorks","Columbia Pictures","Apple Studios","Gaumont","Harpo Productions","Cinemax films","Lucasfilm","Legendary Entertainment"))
        MovieInfo['Director'] = faker.name()
        MovieInfo['Genre'] = faker.random_element(elements=("Action","Comedy","Drama","Horror","Sci-Fi","Fantasy","Romance","Thriller","Mystery"))
        MovieInfo['Box Office Length (Weeks)'] = faker.random_int(min = 4, max = 12) 
        MovieInfo['Budget'] = faker.random_int(min = 10000000, max = 200000000)
        MovieInfo['Revenue'] = faker.random_int(min = 10000000, max = 500000000)
        MovieInfo['Origin of Production'] = faker.state()
        MovieInfo['Rotten Tomatoes Score'] = faker.random_int(min = 0, max = 100)
        MovieInfo['International Release'] = faker.random_element(elements=("Yes","No"))
        movieInformation.append(MovieInfo)
    return pd.DataFrame(movieInformation)
movieNames = generateData(100000)
print(movieNames)
```

```python
#Employee Table Generation
from faker import Faker
import pandas as pd
faker = Faker(locale = 'en_US')
def generateData(num):
    movieInformation =[]
    for i in range(num):
        MovieInfo = {}
        MovieInfo['Shift Time'] = faker.random_element(elements = ("Morning","Mid Day","Night"))
        MovieInfo['Employee ID'] = faker.random_int(min = 1000,max = 2000)
        MovieInfo['Employee Type'] = faker.random_element(elements = ("Box Office", "Concession Stand","Manager"))
        MovieInfo['Name'] = faker.name()
        MovieInfo['Customer Rating'] = faker.random_element(elements = ("Good", "Fair","Poor"))
        movieInformation.append(MovieInfo)
    return pd.DataFrame(movieInformation)
movieNames = generateData(65)
movieNames.to_csv("EmployeeData.csv", index = False)

```

```python
#Concession Table Generation
from faker import Faker
from datetime import datetime
import pandas as pd
faker = Faker(locale = 'en_US')
def generateData(num):
    numbers = [
    1243, 1115, 1152, 1973, 1679, 1592, 1934, 1125, 1208, 1943, 
    1803, 1648, 1975, 1347, 1155, 1645, 1429, 1014, 1791, 1800, 
    1419, 1546, 1610, 1698, 1810, 1632, 1211, 1093, 1079, 1082, 
    1600, 1767, 1651, 1554, 1832, 1494]
    movieInformation =[]
    for i in range(num):
        MovieInfo = {}
        MovieInfo['Concession Type'] = faker.random_element(elements = ("popcorn","soda","candy"))
        if MovieInfo['Concession Type'] == 'popcorn':
            MovieInfo['Concession ID'] = 200
            MovieInfo['Price of Item'] = 7
        if MovieInfo['Concession Type'] == 'soda':
            MovieInfo['Concession ID'] = 201
            MovieInfo['Price of Item'] = 6
        if MovieInfo['Concession Type'] == 'candy':
            MovieInfo['Concession ID'] = 202
            MovieInfo['Price of Item'] = 4
        MovieInfo["Amount Purchased"] = faker.random_int(min = 0, max = 6)
        start_date = datetime.strptime('2023-01-01', '%Y-%m-%d').date()
        end_date = datetime.strptime('2024-01-01', '%Y-%m-%d').date()   
        MovieInfo['Purchased Date'] = faker.date_between(start_date=start_date, end_date=end_date)
        MovieInfo['Employee ID'] = faker.random_element(elements = numbers)
        movieInformation.append(MovieInfo)
    return pd.DataFrame(movieInformation)
movieNames = generateData(45000)
movieNames.to_csv("ConcessionData.csv", index = False)
```

```python
#Customer Table
import numpy as np
from faker import Faker
import pandas as pd
faker = Faker(locale = 'en_US')
def generateData(num):
    movieInformation = []
    customer_id = 4000
    mean = 25
    std_dev = 10
    for i in range(num):
        MovieInfo = {}
        MovieInfo['Customer ID'] = customer_id
        MovieInfo['Customer Name'] = faker.name()
        MovieInfo['Gender'] = faker.random_element(elements = ("M","F"))
        age = int(np.random.normal(loc=mean, scale=std_dev)) 
        age = np.clip(age, 2, 90)
        MovieInfo['Age'] = age
        customer_id += 1
        
        movieInformation.append(MovieInfo)
    
    return pd.DataFrame(movieInformation)
movieNames = generateData(20000)
movieNames.to_csv("CustomerData.csv", index = False)
```

```python
#Show Time Review Table In Progress
from faker import Faker
import pandas as pd
faker = Faker(locale = 'en_US')
comments = ["comfortable seats", "uncomfortable seats", "clean theater", "dirty theater", "movie too long", "good length film", "bad plot", "good plot", "good characters" , "bad characters", "movie was predictable", "movie was boring" , "movie was engaging"] 

def generateData(num):
    movieInformation = []
    reviewNum = 1
    for i in range(num):
        MovieInfo = {}
        MovieInfo['Review ID'] = reviewNum
        MovieInfo['Rating'] = faker.random_int(min = 2, max = 90)
        MovieInfo['Feedback'] = faker.random_element(elements = comments)
        MovieInfo['ShowTime ID'] = faker.random_int(min = 1, max = 20000)
        reviewNum += 1
        movieInformation.append(MovieInfo)
    return pd.DataFrame(movieInformation)
movieNames = generateData(16500)
print(movieNames)



    
```
